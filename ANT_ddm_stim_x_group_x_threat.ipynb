```python
#!/Users/swp7563/Anaconda3/envs/hddmEnv/Scripts
import pandas as pd
import matplotlib.pyplot as plt

import hddm
```

```python
data=hddm.load_csv('C:/Users/swp7563/Anaconda3/envs/hddmEnv/data/ant_ddm.csv')

data.head(10)
```

```python
model_stim_x_threat_x_group = hddm.HDDM (data, depends_on={'v': ['stim', 'threat', 'group']},p_outlier=.05) 
model_stim_x_threat_x_group.find_starting_values()
model_stim_x_threat_x_group.sample(2000, burn=20)
```

```python
v_HC_NS_congruent, v_HC_NS_incongruent, v_HC_SH_congruent, v_HC_SH_incongruent,  v_PDNA_NS_congruent, v_PDNA_NS_incongruent, v_PDNA_SH_congruent, v_PDNA_SH_incongruent, v_PDA_NS_congruent, v_PDA_NS_incongruent, v_PDA_SH_congruent, v_PDA_SH_incongruent= model_stim_x_threat_x_group.nodes_db.node[['v(HC.congruent.NS)', 'v(HC.incongruent.NS)', 'v(HC.congruent.SH)', 'v(HC.incongruent.SH)', 'v(PDNA.congruent.NS)', 'v(PDNA.incongruent.NS)', 'v(PDNA.congruent.SH)', 'v(PDNA.incongruent.SH)', 'v(PDA.congruent.NS)', 'v(PDA.incongruent.NS)', 'v(PDA.congruent.SH)', 'v(PDA.incongruent.SH)']]
```

```python
hddm.analyze.plot_posterior_nodes([v_HC_NS_congruent, v_HC_NS_incongruent, v_HC_SH_congruent, v_HC_SH_incongruent])
plt.xlabel('drift-rate')
plt.ylabel('Posterior probability')
plt.title('Posterior of drift-rate means')
plt.savefig('hddm_HC_stim_x_threat.pdf')
```
